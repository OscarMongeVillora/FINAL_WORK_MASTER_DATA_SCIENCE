```python
%%capture 

from datetime import datetime
start=datetime.now()


#DATASET WORLDWIDE  222 samples
#list_methods_ml = ['xgb', 'SARIMA', 'rf']
list_methods_ml = ['xgb', 'rf', 'SARIMA', 'NN']

rmse_curve_method = {}
for method in list_methods_ml:
    config_template.modify_options(ML_model_choose = method)

        

    rmse_curve = []
    #list_percentage_train = [0.7, 0.8]
    list_percentage_train = [0.4, 0.5, 0.6, 0.7, 0.75, 0.8, 0.85, 0.9]
    #list_percentage_train = [0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
    
    for percentage_train in list_percentage_train:
        size_train = round(percentage_train * 222)
        if method == 'NN':
            config_template.modify_options(n_sample_subset = size_train, bayes_tuning_on = False)
            
        config_template.modify_options(size_train = size_train)
        rmse_value = main_project(config_object=config_template)
        rmse_curve.append(rmse_value)
        
    rmse_curve_method[method] =rmse_curve

```

```python
#print(datetime.now()-start)
fig, ax = plt.subplots(figsize=(9,5))
for method in list_methods_ml[:-1]:
    ax.plot(list_percentage_train, rmse_curve_method[method], '-o', label = method)
ax.set_title('RSMSE vs TRAIN/TEST PERCENTAGE (FOR DIFFERENT ML METHODS')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('RMSE ERROR in predictions')
plt.legend()
plt.show()
```

```python
#PLOT THE RSME CURVE FOR DIFFERENT MODELS (DIFFERENT PROPORTIONS TRAIN/TEST)

print(finish44-start44)
fig, ax = plt.subplots(figsize=(9,5))
for method in list_methods_ml:
    ax.plot(list_percentage_train44, rmse_curve_method44[method], '-o', label = method)
ax.set_title('RSMSE vs TRAIN/TEST PERCENTAGE (FOR DIFFERENT ML METHODS)')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('RMSE ERROR in predictions')
plt.legend()
plt.show()
```

```python
#COMPARATION GOMPERTZ VS NULL (WON'T INCLUDE NN) CUM

fig, ax = plt.subplots(figsize=(10,6))
colour_list =['royalblue', 'darkorange', 'seagreen']
for method, colour in zip(list_methods_ml, colour_list) :
    ax.plot(list_percentage_train11, rmse_curve_method11[method], '-o', c = colour, label = method)
    ax.plot(list_percentage_train44, rmse_curve_method44[method], '--o', c= colour, label = method + ' (NO TREND FIT)')
ax.set_title('RSMSE vs TRAIN/TEST PERCENTAGE (GOMPERTZ FIT VS NONE)')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('RMSE ERROR in predictions')
plt.legend()
plt.show()

```

```python
%%capture
#BEST PREDICTIONS



 
from datetime import datetime
start=datetime.now()


#DATASET WORLDWIDE  222 samples
#list_methods_ml = ['xgb', 'SARIMA', 'rf']
list_methods_ml1_4 = ['xgb', 'rf', 'SARIMA']

data_curve_method1_4 = {}
rmse_curve_method1_4 = {}
for method in list_methods_ml1_4:
    config_template.modify_options(ML_model_choose = method)
    data_curve = [] 
    size_train = round(0.9 * 222)


    config_template.modify_options(size_train = size_train)
    rmse_value, data_out = main_project(config_object=config_template)
    data_train = data_out[0]
    data_test = data_out[1]

    rmse_curve_method1_4[method] =rmse_value
    data_curve_method1_4[method] =data_out[3]


finish = datetime.now()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(range(data_train.shape[1]),data_train.squeeze(), label = 'train data')
ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), data_test.squeeze(), label = 'test data')

for method in list_methods_ml1_4 :
    error_test = rmse_curve_method1_4[method]
    predict_test = data_curve_method1_4[method]
    ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), 
            predict_test.squeeze() ,label = method +' prediction (RMSE = ' +  str(error_test)[:-8] +')')

ax.set_title('TEST PREDICTIONS FOR DIFFERENT ML MODELS')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('RMSE ERROR in predictions')
plt.legend()
plt.show()
```
