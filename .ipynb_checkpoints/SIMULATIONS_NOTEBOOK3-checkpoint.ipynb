```python
#%matplotlib notebook
from config_generator import config_generator
from main_project import main_project
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
#ALL 80 ITER DEATHS WORLWIDE
%matplotlib notebook
```

```python
# 1) DEFINE A CONFIGURATION TEMPLATE: DATASET DEATHS_UPDATE (, CUMULATIVE MODE, GOMPERTZ, ITERATIONS_BAYES = 80
config_template = config_generator()
config_template.import_options_template(name_dataset = 'deaths_update', mode_dataset = 'diff', method_trend = 'gompertz',
                            ML_model_choose = 'xgb', predict_intervals = False, size_train = 64,
                            list_comunities =['Cataluña', 'Madrid'], all_comunities_bool = False, n_sample_subset = 28,
                            loss_function = 0, timesplitmethod = 0, percentage_validation =0.3, overlap = 0.3,
                            max_iter = 80, plot_mode = 2, metrics_print = True, plot_print = False, 
                            bayes_tuning_on = True) 

#Run the simulation  (Replace the configuration options)
rmse, data = main_project(config_object = config_template)
```

```python
%%capture 
#CALCULATE RMSE PREDICTIONS FOR DIFFERENT MODELS FOR DIFFERENT PROPORTIONS OF TRAIN/TEST
start1=datetime.now()

#DATASET WORLDWIDE  222 samples
list_methods_ml = ['xgb', 'rf', 'SARIMA']

rmse_curve_method1 = {}
data_curve_method1_07 = {}
data_curve_method1_085 = {}

for method in list_methods_ml:
    config_template.modify_options(ML_model_choose = method)
    rmse_curve = []
    list_percentage_train = [0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
    if method == 'SARIMA':
        config_template.modify_options(max_iter = 50)
    
    for percentage_train in list_percentage_train:
        #RMSE TAKEN
        size_train = round(percentage_train * 80)            
        config_template.modify_options(size_train = size_train)
        rmse_value, data_out = main_project(config_object=config_template)
        rmse_curve.append(rmse_value)
        
        #PREDICTIONS TAKEN
        if percentage_train == 0.7:
            data_curve_method1_07[method] =data_out
        if percentage_train == 0.85:
            data_curve_method1_085[method] =data_out
            
    rmse_curve_method1[method] =rmse_curve

finish1 = datetime.now()
```

```python
#PRINT ERRORS

fig, ax = plt.subplots(figsize=(9,5))
for method in list_methods_ml:
    ax.plot(list_percentage_train, rmse_curve_method1[method], '-o', label = method)
ax.set_title('RSMSE vs TRAIN/TEST PERCENTAGE (FOR DIFFERENT ML METHODS)')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('RMSE ERROR in predictions')
plt.legend()
plt.show()
```

```python
#PREDICTION IN 0.85 TRAIN/TEST

data_train = data_curve_method1_085['xgb'][0]
data_test = data_curve_method1_085['xgb'][1]

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(range(data_train.shape[1]),data_train.squeeze(), label = 'train data')
ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), data_test.squeeze(), lw = 3, label = 'test data')

for method in list_methods_ml :
    error_test = rmse_curve_method1[method][2]
    data_train = data_curve_method1_085[method][0]
    data_test = data_curve_method1_085[method][1]
    predict_test = data_curve_method1_085[method][3]
    ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), 
            predict_test.squeeze() ,label = method +' prediction (RMSE = ' +  str(error_test)[:-8] +')')

ax.set_title('TEST PREDICTIONS FOR DIFFERENT ML MODELS')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('COVID-19 Cataluña + Madrid Daily deaths')
plt.legend()
plt.show()
```

```python
#PREDICTION IN 0.7 TRAIN/TEST

data_train = data_curve_method1_07['xgb'][0]
data_test = data_curve_method1_07['xgb'][1]

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(range(data_train.shape[1]),data_train.squeeze(), label = 'train data')
ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), data_test.squeeze(), lw = 3, label = 'test data')

for method in list_methods_ml :
    error_test = rmse_curve_method1[method][2]
    data_train = data_curve_method1_07[method][0]
    data_test = data_curve_method1_07[method][1]
    predict_test = data_curve_method1_07[method][3]
    ax.plot(range(data_train.shape[1], data_train.shape[1] + data_test.shape[1]), 
            predict_test.squeeze() ,label = method +' prediction (RMSE = ' +  str(error_test)[:-8] +')')

ax.set_title('TEST PREDICTIONS FOR DIFFERENT ML MODELS')
ax.grid()
plt.xlabel('Train Percentage of Dataset')
plt.ylabel('COVID-19 Cataluña + Madrid Daily deaths')
plt.legend()
plt.show()
```
